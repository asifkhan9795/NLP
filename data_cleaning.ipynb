```python
# prompt: create sample dataset with urls , hashtags

import pandas as pd

# Create a sample dataset with URLs and hashtags
data = {
    'text': [
        "Check out this amazing article! https://www.example.com #news #technology",
        "Just watched a great movie. #movies #entertainment",
        "Excited for the new season! Visit our website: www.website.com #sports #football",
        "Having fun at the beach! #travel #vacation",
        "This is a great product. https://shop.example.com #shopping #deals"
    ]
}

df_sample = pd.DataFrame(data)

print(df_sample)
```

```python
# prompt: create a funtion that can remove URL, hashtags from the DF (df_sample)

import re

def remove_urls_hashtags(df, text_column):
  """
  Removes URLs and hashtags from a specified column in a Pandas DataFrame.

  Args:
    df: The Pandas DataFrame.
    text_column: The name of the column containing the text.

  Returns:
    The DataFrame with URLs and hashtags removed from the specified column.
  """

  df[text_column] = df[text_column].apply(lambda text: re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE))
  df[text_column] = df[text_column].apply(lambda text: re.sub(r'#\w+', '', text))
  return df


# Example usage:
df_cleaned = remove_urls_hashtags(df_sample, 'text')
print(df_cleaned)
```

# Text Preprocessing on Spam Text Message

```python
# Loading dataset
import pandas as pd

df = pd.read_csv('/content/SPAM text message 20170820 - Data.csv')
df.head()
```

# Removing URL, Hashtags, Symbols, Emoji

```python
# prompt: Using dataframe df: write a regex that can clean URLs, Links, hashtags, Emoji,

import re

def clean_text(text):
  """
  This function cleans the text by removing URLs, links, hashtags, emojis, and other unwanted characters.
  """
  text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Remove URLs
  text = re.sub(r'#\w+', '', text)  # Remove hashtags
  text = re.sub(r'[^\x00-\x7F]+', '', text)  # Remove emojis and other non-ASCII characters
  text = re.sub(r'\s+', ' ', text).strip() # Remove extra whitespace
  return text

# Example usage: You can apply this function to your 'Message' column
df['cln_URL_#_Emo'] = df['Message'].apply(clean_text)
```

```python
df.head()
```

# Converting data to lower case

```python
# prompt: Using dataframe df: now converty the column cln_URL_#_Emo to lower case and save it int new column

# Convert the 'cln_URL_#_Emo' column to lowercase.
df['cln_URL_#_Emo_lower'] = df['cln_URL_#_Emo'].str.lower()
df.head()
```

# Tokinization of data

```python
# prompt: Using dataframe df: now tokenize the column cln_URL_#_Emo_lower and create new column

# Import necessary library
from nltk.tokenize import word_tokenize

# Create a new column 'tokenized_text' by tokenizing the 'cln_URL_#_Emo_lower' column
df['tokenized_text'] = df['cln_URL_#_Emo_lower'].apply(lambda text: word_tokenize(text))
```

# Applying lammatization

```python
# prompt: Using dataframe df: now lammatize the column tokenized_text and create new column

# Import necessary libraries
import nltk
from nltk.stem import WordNetLemmatizer

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Function to lemmatize a list of tokens
def lemmatize_tokens(tokens):
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return lemmatized_tokens

# Apply the lemmatize_tokens function to the 'tokenized_text' column and create a new column
df['lemmatized_tokenized_text'] = df['tokenized_text'].apply(lemmatize_tokens)
```

```python
df.head()
```

# Applying Stemmer

```python
# prompt: Using dataframe df: now apply stemmer on column lemmatized_tokenized_text and create new column

# Import the necessary library for stemming
from nltk.stem import PorterStemmer

# Create a stemmer object
stemmer = PorterStemmer()

# Apply stemming to each token in the lemmatized_tokenized_text column
# and create a new column named stemmed_text
df['stemmed_text'] = df['lemmatized_tokenized_text'].apply(lambda tokens: [stemmer.stem(token) for token in tokens])
df.head()
```

# Removing Stop words

```python
# prompt: Using dataframe df: now remove stopswords from column stemmed_text and create a new column

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Function to remove stopwords from a list of tokens
def remove_stopwords(tokens):
  if isinstance(tokens, list):
    stop_words = set(stopwords.words('english'))
    return [word for word in tokens if word.lower() not in stop_words]
  else:
    return []

# Apply the function to the 'stemmed_text' column and create a new column 'stemmed_text_no_stopwords'
df['stemmed_text_no_stopwords'] = df['stemmed_text'].apply(remove_stopwords)
df.head()
```
